```python
# Group the data by city and calculate the sum of gross income for each city
city_income = df.groupby('City')['gross income'].sum().reset_index()

# Sort the cities by gross income in descending order
city_income = city_income.sort_values(by='gross income', ascending=False)

# Get the top cities with the highest gross income
top_cities = city_income.head()

# Print the top cities with their gross income
print("Top Cities by Gross Income:")
print(top_cities)
```

```python
# Group the data by City and Branch and calculate the sum of Gross Income
city_branch_income = df.groupby(['City', 'Branch'])['gross income'].sum().reset_index()

# Sort the cities by gross income in descending order
city_branch_income = city_branch_income.sort_values(by='gross income', ascending=False)

# Get the top 5 cities with the highest gross income, considering the branch
top_cities = city_branch_income.head()

# Print the top cities with their gross income and branch
print("Top Cities by Gross Income and Branch:")
print(top_cities)
```

```python
# Group the data by City and Branch and calculate the sum of Gross Income
city_branch_income = df.groupby(['gross income', 'Branch'])['City'].sum().reset_index()

# Sort the cities by gross income in descending order
city_branch_income = city_branch_income.sort_values(by='gross income', ascending=False)

# Get the top 10 cities with the highest gross income, considering the branch
top_10_cities = city_branch_income.head(10)

# Print the top 10 cities with their gross income and branch
print("Top 10 Cities by Gross Income and Branch:")
print(top_10_cities)
```

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
sns.boxplot(x='Branch', y='Unit price', data= df, palette='rainbow')
```

```python
# Example: Bar chart to show average gross income by product line for each gender
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

sns.barplot(x="Product line", y="gross income", hue="Gender", data=df)

plt.title("Average Gross Income by Product Line and Gender")
plt.xlabel("Product Line")
plt.ylabel("Average Gross Income")
plt.xticks(rotation=45)

plt.show()
```

```python
# Perform a two-way ANOVA
result = stats.f_oneway(*[df[df['City'] == city]['Tax 5%'] for city in df['City'].unique()],
                       *[df[df['Branch'] == branch]['Tax 5%'] for branch in df['Branch'].unique()])

# Set the significance level (alpha)
alpha = 0.05

# Print the results
print("F-statistic:", result.statistic)
print("P-value:", result.pvalue)

if result.pvalue < alpha:
    print("There is a significant relationship between City, Branch, and Taxes Paid.")
else:
    print("There is no significant relationship between City, Branch, and Taxes Paid.")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Select the relevant columns
selected_columns = ["Total", "Quantity", "Tax 5%", "cogs", "gross income", "Rating"]

# Create a subset of the data with the selected columns
subset_data = df[selected_columns]

# Calculate the correlation matrix
correlation_matrix = subset_data.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Plot")
plt.show()
```
